```python
import numpy as np
import numpy.random as random
import scipy as sp
import pandas as pd
from pandas import Series, DataFrame

# 可視化ライブラリ
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
sns.set()
%matplotlib inline

# 小数第三位まで表示
%precision 3
```

```python
'''
棒グラフ
'''

# 表示するデータ
x = [1, 2, 3]
y = [10, 1, 4]

# グラフの大きさ指定
plt.figure(figsize=(10,6))

plt.bar(x, y, align='center', width=0.5)

# 棒グラフそれぞれのラベル
plt.xticks(x, ['A Class', 'B Class', 'C Class'])

# xとyのラベルを設定
plt.xlabel('Class')
plt.ylabel('Score')

# グリッドを表示
plt.grid(True)
```

```python
'''
横の棒グラフ
'''

# 表示するデータ
x = [1, 2, 3]
y = [10, 1, 4]

# グラフの大きさ指定
plt.figure(figsize=(10,6))

plt.barh(x, y, align='center')

# 棒グラフそれぞれのラベル
plt.yticks(x, ['A Class', 'B Class', 'C Class'])

# xとyのラベルを設定
plt.ylabel('Class')
plt.xlabel('Score')

# グリッドを表示
plt.grid(True)
```

```python
'''
複数のグラフ
'''

# データの準備
y1 = np.array([30, 10, 40])
y2 = np.array([10, 50, 90])

# x軸のデータ
x = np.arange(len(y1))

# グラフの幅
w = 0.4

# グラフの大きさ指定
plt.figure(figsize=(10,6))

# グラフの描画
plt.bar(x, y1, color='blue', width=w, label='Math first', align='center')
plt.bar(x+w, y2, color='green', width=w, label='Math final', align='center')

# 凡例を最適な位置に配置
plt.legend(loc='best')

plt.xticks(x+w /2 , ['A Class', 'B Class', 'C Class'])

# グリッドを表示
plt.grid(True)
```

```python
'''
積み上げ棒グラフ
'''

# データの準備
height1 = np.array([100, 200, 300, 400, 500])
height2 = np.array([1000, 800, 600, 400, 200])

# x軸
x = np.array([1, 2, 3, 4, 5])

# グラフの大きさ指定
plt.figure(figsize=(10,6))

# グラフの描画
p1 = plt.bar(x, height1, color='blue')
p2 = plt.bar(x, height2, bottom=height1, color='lightblue')

# 凡例を最適な位置に配置
plt.legend((p1[0], p2[0]), ('Class1', 'Class2'))
```

```python
'''
円グラフ
'''
labels = ['Frogs', 'Hogs', 'Dogs', 'Logs']
sizes = [15, 30, 45, 10]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0.1, 0, 0)

# グラフの大きさ指定
plt.figure(figsize=(10,6))

# グラフの表示
plt.pie(sizes, explode = explode, labels=labels, colors=colors,
       autopct='%1.1f%%', shadow=True, startangle=90)

# 円を丸く描画
plt.axis('equal')
```

```python
'''
バブルチャート
'''

N = 25

# X, Yデータをランダムに生成
x = np.random.rand(N)
y = np.random.rand(N)

# color番号
colors = np.random.rand(N)

# バブルの大きさをばらけさせる
area = 10 * np.pi * (15 * np.random.rand(N)) ** 2

# グラフの大きさ指定
plt.figure(figsize=(10,6))

# グラフの描画
plt.scatter(x, y, s = area, c = colors, alpha = 0.5)
plt.grid(True)
```

```python
'''
練習問題7-1
'''

# データ取得
mat_df = pd.read_csv("student-mat.csv", sep=';')
pie_df = pd.DataFrame(mat_df['reason'].value_counts())

labels = []
sizes = []
explode_li = []
for index, row in pie_df.iterrows():
    labels.append(index)
    sizes.append(row['reason'])
    explode_li.append(0)
    
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = tuple(explode_li)

# グラフの大きさ指定
plt.figure(figsize=(10,6))

# グラフの表示
plt.pie(sizes, explode = explode, labels=labels, colors=colors,
       autopct='%1.1f%%', shadow=True, startangle=90)

# 円を丸く描画
plt.axis('equal')
```

```python
'''
練習問題7-2
'''

# データ取得
mat_df = pd.read_csv("student-mat.csv", sep=';')
calc_df = pd.DataFrame(mat_df.groupby('higher', as_index=False).mean()[['higher', 'G3']])

# 表示するデータ
x = [1, 2]
y = list(calc_df['G3'])

# グラフの大きさ指定
plt.figure(figsize=(10,6))

plt.bar(x, y, align='center', width=0.5)

# 棒グラフそれぞれのラベル
plt.xticks(x, ['no', 'yes'])

# xとyのラベルを設定
plt.xlabel('higher')
plt.ylabel('G3 Score Average')

# グリッドを表示
plt.grid(True)
```

```python
'''
練習問題7-3
'''

# データ取得
mat_df = pd.read_csv("student-mat.csv", sep=';')
calc_df = pd.DataFrame(mat_df.groupby('traveltime', as_index=False).mean()[['traveltime', 'G3']])

# 表示するデータ
x = list(calc_df['traveltime'])
y = list(calc_df['G3'])

# グラフの大きさ指定
plt.figure(figsize=(10,6))

plt.barh(x, y, align='center')

# 棒グラフそれぞれのラベル
plt.yticks(x, ['1', '2', '3', '4'])

# xとyのラベルを設定
plt.ylabel('traveltime')
plt.xlabel('G3 Score')

# グリッドを表示
plt.grid(True)
```

```python
# 日付データの設定。freq='T'で1分ごとにデータを生成する
idx = pd.date_range('2015/01/01', '2015/12/31 23:59', freq='T')

# 乱数の発生。1か-1を発生させる
dn = np.random.randint(2, size=len(idx)) * 2 - 1

# ランダムウォークを作成
rnd_walk = np.cumprod(np.exp(dn * 0.0002)) * 100

# resample('B')でデータを営業日単位でリサンプリング
df = pd.Series(rnd_walk, index=idx).resample('B').ohlc()
```
